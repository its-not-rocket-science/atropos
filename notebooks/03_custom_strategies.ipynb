# Custom Optimization Strategies

Learn how to define and evaluate your own pruning strategies.

```python
import pandas as pd

from atropos import OptimizationStrategy, estimate_outcome
from atropos.presets import SCENARIOS, STRATEGIES
```

## Define Custom Strategies

```python
# Conservative for quality-critical applications
conservative = OptimizationStrategy(
    name="conservative",
    parameter_reduction_fraction=0.10,
    memory_reduction_fraction=0.08,
    throughput_improvement_fraction=0.05,
    power_reduction_fraction=0.04,
    quality_risk="low",
)

# Aggressive for cost-sensitive applications
aggressive = OptimizationStrategy(
    name="aggressive",
    parameter_reduction_fraction=0.50,
    memory_reduction_fraction=0.40,
    throughput_improvement_fraction=0.35,
    power_reduction_fraction=0.25,
    quality_risk="high",
)

custom_strategies = [conservative, aggressive]
```

## Evaluate Custom Strategies

```python
scenario = SCENARIOS["medium-coder"]

results = []
for strategy in custom_strategies:
    outcome = estimate_outcome(scenario, strategy)
    mem_red = strategy.memory_reduction_fraction * 100
    results.append(
        {
            "Strategy": strategy.name,
            "Memory Reduction %": mem_red,
            "Annual Savings ($)": outcome.annual_total_savings_usd,
            "Risk": strategy.quality_risk,
        }
    )

pd.DataFrame(results)
```

## Compare with Built-ins

```python
all_strategies = list(STRATEGIES.values()) + custom_strategies

results = []
for strategy in all_strategies:
    outcome = estimate_outcome(scenario, strategy)
    be = outcome.break_even_years
    be_months = be * 12 if be else None
    results.append(
        {
            "Strategy": strategy.name,
            "Annual Savings ($)": outcome.annual_total_savings_usd,
            "Break-even (months)": be_months,
        }
    )

pd.DataFrame(results)
```
